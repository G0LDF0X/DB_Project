```python
import pandas as pd
import numpy as np
import os
import csv
import random
```

```python
cwd = os.getcwd()
file_path_1 = os.path.join(cwd, 'male.txt')
file_path_2 = os.path.join(cwd, 'female.txt')
```

```python
with open(file_path_1, encoding='utf-8') as file1:
    data1 = file1.read()
    data1 = data1.replace('\n', '')
    data1 = data1.split('<td class="">')
    
    with open(file_path_2, encoding='utf-8') as file2:
        data2 = file2.read()
        data2 = data2.replace('\n', '')
        data2 = data2.split('<td class="">')
```

```python
male_name = []
female_name = []
first_name = ['김', '이', '박', '최', '정', '강', '조', '윤', '장', '임', '한', '서']

for i in range(0, len(data1)):
    if 'name' in data1[i]:
        tmp = data1[i]
        if tmp[100] != '\"':
            tmp = tmp[99:101]
        else:
            tmp = tmp[99:100]
        male_name.append(tmp)
        
for i in range(0, len(data1)):
    if 'name' in data2[i]:
        tmp = data2[i]
        if tmp[100] != '\"':
            tmp = tmp[99:101]
        else:
            tmp = tmp[99:100]
        female_name.append(tmp)
```

```python
major = []
student_total_number = []
f = open('major.csv', 'r', encoding='utf-8-sig')
reader = csv.reader(f)
header = next(reader)

for line in reader:
    major.append(line[0])
    student_total_number.append(line[1])
f.close()
```

```python
subject_code = []
subject_grade = []
subject_major = []
subject = []
tmp_list = []
f = open('final_db.csv', 'r', encoding = 'utf-8-sig')
rdr = csv.reader(f)
header = next(rdr)

for line in rdr:
    if line[9] not in subject_code :
        subject_code.append(line[9])
        subject_grade.append(line[7])
        subject_major.append(line[10])

for i in range(0, len(subject_code)):
    tmp_list = []
    tmp_list.append(subject_code[i])
    tmp_list.append(subject_grade[i])
    tmp_list.append(subject_major[i])
    subject.append(tmp_list)
    
f.close()

subject_all = ['교직', '교양', '']
```

```python
def line_processing(line):
    ret = {}
    ret['student_name'] = np.NaN
    ret['student_id'] = np.NaN
    ret['major'] = np.NaN
    ret['period'] = np.NaN
    ret['grade'] = np.NaN
    ret['passed_subject'] = np.NaN
    ret['get_subject'] = np.NaN
    get_sub = []
    passed_sub = []
    
    if random.randint(1, 2) == 1:
        ret['student_name'] = first_name[random.randint(0, len(first_name) - 1)] + male_name[random.randint(0, len(male_name) - 1)]
    else:
        ret['student_name'] = first_name[random.randint(0, len(first_name) - 1)] + female_name[random.randint(0, len(female_name) - 1)]
    
    for i in range(0, len(student_total_number)):
        if line < int(student_total_number[i]):
            ret['major'] = major[i]
            break
    
    ret['period'] = random.randint(1, 8)
    
    if ret['period'] <= 2 :
        ret['grade'] = 1
    elif ret['period'] <= 4:
        ret['grade'] = 2
    elif ret['period'] <= 6:
        ret['grade'] = 3
    else:
        ret['grade'] = 4

    if(ret['grade'] == 1):
        ret['student_id'] = "2021" + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9))
    elif(ret['grade'] == 2):
        ret['student_id'] = "20" + str(random.randint(11, 20)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9))
    elif(ret['grade'] == 3):
        ret['student_id'] = "20" + str(random.randint(11, 19)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9))
    else:
        ret['student_id'] = "20" + str(random.randint(11, 18)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9)) + str(random.randint(0, 9))
    
    length = random.randint(4, 6)
    while len(get_sub) < length:
        i = 0
        tmp_sub = subject[random.randint(0, len(subject) - 1)]
        while(True):
            if(i > len(subject) - 1):
                break
            if(tmp_sub[0] == subject[i][0]):
                if(ret['grade'] == subject[i][1] or subject[i][1] == ''):
                    if(ret['major'] in subject[i][2] or subject[i][2] in subject_all):
                        if tmp_sub[0] not in get_sub:
                            get_sub.append(tmp_sub[0])
                        break
            i += 1
        ret['get_subject'] = get_sub
    
    
    for k in range(0, ret['grade'] * random.randint(4, 6)) :
        check = subject[random.randint(0, len(subject) - 1)]
        if check[0] not in ret['get_subject']:
            passed_sub.append(check[0])
    ret['passed_subject'] = passed_sub
    
    return ret
```

```python
from IPython.display import display
processed_datas = []
keys = ['student_name', 'student_id', 'major', 'period', 'grade', 'passed_subject', 'get_subject']

df = pd.DataFrame(columns = keys)

for i in range(0, 4784):
    processed_data = line_processing(i)
    df = df.append({
        'student_name' : processed_data['student_name'],
        'student_id' : processed_data['student_id'],
        'major' : processed_data['major'],
        'period' : processed_data['period'],
        'grade' : processed_data['grade'],
        'passed_subject' : processed_data['passed_subject'],
        'get_subject' : processed_data['get_subject']
    }, ignore_index=True)
    
df.to_csv('random_student.csv', encoding='utf-8-sig')
```
